```python
import pandas as pd 
```

```python
df = pd.read_csv("../resources/global_superstore.csv")
df
```

```python
df.columns
```

```python
customer_df = df.groupby(['Customer ID']).first().reset_index()[['Customer ID', 'Customer Name', 'Segment', 'City', 'State', 'Country',
       'Postal Code', 'Market', 'Region']]

customer_df
# customer_df["Postal Code"] = customer_df["Postal Code"].astype(int).astype(str)
customer_df.head()
customer_df.to_csv("../resources/customers.csv", index=False)
```

```python
orders_df=df.groupby(['Order ID']).first().reset_index()[['Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
       'Customer ID','Order Priority']]

orders_df.to_csv("../resources/orders.csv",index=False)
```

```python
orders_df
```

```python
order_row_df = df.groupby(['Order ID', 'Product ID', 'Category',
       'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount',
       'Profit', 'Shipping Cost']).count().reset_index()[['Order ID', 'Product ID', 'Category',
       'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount',
       'Profit', 'Shipping Cost']]

order_row_df[order_row_df["Order ID"]=="MX-2012-155047"]
```

```python
order_row_df['Order Item Number'] = order_row_df.groupby(['Order ID']).cumcount()+1
order_row_df
```

```python
order_row_df["Order Item ID"] = order_row_df["Order ID"] + "-" + order_row_df["Order Item Number"].astype(str)
order_row_df.head()
order_row_df = order_row_df[['Order Item ID', 'Order ID', 'Order Item Number', 'Product ID', 'Category',
       'Sub-Category', 'Product Name', 'Sales', 'Quantity', 'Discount',
       'Profit', 'Shipping Cost']]

order_row_df.head()
```
